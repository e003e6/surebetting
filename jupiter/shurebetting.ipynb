```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def megeri(o1, o2):
    return 1/o1 + 1/o2
```

```python
def szamitas(o1, o2, x1, x2):
    t = x1+x2
    kif1 = o1*x1 - t
    kif2 = o2*x2 - t 
    print(megeri(o1, o2))
    print(kif1, kif2)
```

```python
o1 = np.arange(1, 5, 0.1)
o2 = np.arange(1, 5, 0.1)
```

```python
for i in range(len(o1)):
    for j in range(len(o1)):
        print('odds 1:', o1[i], '\t odds 2:', o2[j], '\t', megeri(o1[i], o2[j]))
```

```python
szamitas(1.2, 2, 50000, 50000)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X, Y = np.meshgrid(o1, o2)

Z = megeri(X, Y)
Z_masked = np.ma.masked_where(Z >= 1, Z)

plt.contourf(X, Y, Z_masked, levels=50, cmap='plasma')
plt.colorbar(label='f(o1, o2) = 1/o1 + 1/o2')
plt.xlabel("o1")
plt.ylabel("o2")
plt.title("Profitbilis f(o1,o2) < 1")
plt.axis('equal')
plt.show()
```
