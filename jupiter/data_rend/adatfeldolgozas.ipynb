```python
meccs_id = 'venezuela-kolumbia-25-09-08'
```

```python
keys = r.keys(f"{meccs_id}*")
values = r.mget(keys)
data = {k.removeprefix(f"{meccs_id}-"): json.loads(v) for k, v in zip(keys, values)}
```

```python
for iroda, fogadasok in data.items():
    print(iroda)
    for fogadas, oddsok in fogadasok.items(): 
        if len(oddsok) < 2 or len(oddsok) > 3:
            continue
        print('\t', fogadas)
        print('\t\t', oddsok)
        print('\t\t', len(oddsok))
        print()
```
