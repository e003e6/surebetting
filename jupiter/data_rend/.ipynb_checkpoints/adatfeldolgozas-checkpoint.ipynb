```python
def get_parok(kulcsok):
    groups = defaultdict(list)
    
    for k in kulcsok:
        if k == "lastupdate":
            continue
    
        prefix = "-".join(k.split("-")[:4])
        groups[prefix].append(k)
    
    pairs = [tuple(v) for v in groups.values() if len(v) > 1]

    return pairs
```
